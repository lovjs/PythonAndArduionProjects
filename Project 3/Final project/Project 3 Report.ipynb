# Exercise 1:

For part one our goal was to use the provided example to estimate the number of times that we expect to see three or more defects in less than ten minutes in an eight hour shift.  To do this we just need to make a list that has the amount of times that there were three defects and see when that list has a value thats < 10.  

```python

Tau = 10.0
t = -Tau*np.log(1.0-np.random.rand(48))  # get 48 intervals between defects

delta2 = t[:-1] + t[1:]
ThreeInLessThan10 = delta2 < 10



print("Number of times 3 defects happened in less than 10 minutes:", len(delta2[np.nonzero(ThreeInLessThan10)]))
```

# 6 Statistics

# Exercise 2:

For part two I changed final time to 480 since its eight hours. Then I inserted and if statement for when currTime > 300 because the last three hours theres an 10% increase in defects so I changed processTime to 9. I got 48 defects which makes since because six per hour for eight hours is 48. However this value will probably change when you run it.




```python
#
# Let's simulate an assembly line for one hour as an example.
#

finalTime = 480            # run for one hour
processTime = 10.0        # how long does it take to process an 
currTime = 0.0            # what time is it now?
nextTime = processTime    # when will the last defect be finished? (it takes 10 minutes to handle a defect)
defectsProcessed = 0      # how many defects have been processed?
defectsOccurred = 0

for delta in t:
    currTime = currTime + delta # when does the next defect appear?

    if currTime > finalTime:
        break # OK, it's the end of the 8 hour shift, break out.

    defectsOccurred += 1  # count the number of defects that have happened.

    if currTime > nextTime:
        nextTime = currTime + processTime # process immediately
        
    if currTime > 300:
        processTime = 9.0
    else:
        nextTime = nextTime + processTime # process after this one.
        
    if nextTime<finalTime:
        defectsProcessed += 1
        
print("Defects Occurred:", defectsOccurred)
print("Defects Processed:", defectsProcessed)
print("Backlog:", defectsOccurred-defectsProcessed)
```
