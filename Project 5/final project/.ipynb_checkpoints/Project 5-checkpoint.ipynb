# 4 Code

For this week the project was really centered around the code. That said there actually isn't much thats new to discus. All the magic is in this one line. Essentially this line stops whatever process is currently happening and says to the arduino I'M THE MOST IMPORTANT!!!!!! Then the arduino executes whatever code that the interrupt demands and then goes back to its business. 
![Interupt code picture](code.png)


#### Code
volatile long int count = 0L;

void setup()
{
  pinMode(13, OUTPUT);
  Serial.begin(9600);
  attachInterrupt(digitalPinToInterrupt(2), click, FALLING);
}

void click() {
  count++;
  digitalWrite(13, ((count>>3) & 0x01));
}

void loop()
{
  delay(1000); // Wait for 1000 millisecond(s)
  Serial.println(count);
  count=0L;
}

# Exercise 1:

The goal of this exercise was to change the value of lambda to the mean of your data and see how that effects the graph. So I did just that and the graph looks beautiful!!


```python
df = pd.read_excel("geigerData.xlsx")
x = df.Counts_per_second.values
df
```

```python
lam= x.mean()  # pick a mean value for the distribution
hrange=np.arange(0,2*lam)
nd = norm.pdf(hrange,loc=lam,scale=np.sqrt(lam)) # plot the normal distribution
pl.plot(hrange, poisson.pmf(hrange, lam), 'r+')  # plot the poisson pmf alongside
pl.plot(hrange,nd,'b-')
pl.title("Compare poisson and normal")
pl.xlabel("how many defects?")
pl.ylabel("probability of defects")
pl.grid()
```
