 int flag = 0;        // set when data is collected
const int N = 150;   // how many data points?

int v1[N];// raw analogRead vals
int v2[N];
long times[N];       // times (microseconds)

void setup()
{
  pinMode(13, OUTPUT);
  pinMode(3,OUTPUT);
  Serial.begin(9600);
  analogWrite(3,51);
}

/*
** collectData iterates through the arrays and stores
** times and raw voltages as quickly as possible
*/

void collectData() {
  for (int i=0; i<N; i++) {
    v1[i]=analogRead(0);
    v2[i]=analogRead(1);
    times[i]=micros();
  }
}

/*
** printData iterates through the arrays and prints the
** stored values back to the computer.
*/

void printData() {
  Serial.print("Time:");
  Serial.print("\tVoltage 1:");
  Serial.println("\tVoltage 2:");
  for (int i=0; i<N; i++) {
    Serial.print(times[i]);
    Serial.print("\t");
    Serial.print(v1[i] * (5.0 / 1023.0));
    Serial.print("\t\t");
    Serial.println(v2[i] * (5.0 / 1023.0));
  }
}

void loop()
{
  if (flag==0) {
    collectData();
    printData();
    flag=1;
  } else {
    /*
    ** We got the data, just flash the LED now.
    */
    digitalWrite(13, HIGH);
    delay(1000); // Wait for 1000 millisecond(s)
    digitalWrite(13, LOW);
    delay(1000); // Wait for 1000 millisecond(s)
  }
}

 # 5 Analysis
 
 # Exercise 1:
 
 So our goal here is to get the value of beta. To do this I started by using ohms law to find the currents. So I took the voltages form before and after R2 to get Ib and then used voltage from before R3 to get Ic. I wanted these values so I could use the equation $$ I_{ce} = \beta I_{be} $$
 Then I took only those values and graphed them in order to find the slope which is equal to beta. I suspected to get a value around 20ish, however I got a value of 200 which is much higher than I expected. I think it was higher due to my resistor values.   

```python
plt.figure(figsize = (10,4))
R2 = 8200
R3 = 500
Ib = (data['Befr_R2:'] - data['Aftr_R2:']) / R2
Ic = (5.0 - data['Befr_R3:']) / R3
x = Ib[Ib < 0.0002]
y = Ic[Ib < 0.0002]
plt.plot(x,y,'b.')
plt.grid()
plt.title('Collector current over Base current')
plt.xlabel('Ib (amps)')
plt.ylabel('Ic (amps)') 
plt.legend()
```

```python
betax = Ib[Ib < .00005] 
betay = Ic[Ib < .00005] 
plt.plot(betax, betay, 'b.')
plt.title('Collector current over Base current while Ib < .00005')
plt.xlabel('Ib (amps)')
plt.ylabel('Ic (amps)') 
plt.legend()

slope = (np.mean(betay)) / (np.mean(betax))

print('Beta is roughly equal to the slope which is ', slope)
```

 # Exercise 2:
 
 For this exercise our goal was to design a circuit that will produce a gain of two using the LM324 with a sawtooth waveform as an input. To verify that our actual gain is two I needed to use the equation $$ G = \frac{Vout}{Vin} $$ So I took the voltages we collected and graphed Voltage 1 over Voltage 2. Then I found the slope of that and bingo!! It was two volts, which means that we wired our circuit correctly. 

```python
plt.figure(figsize = (10,4))
vout = data2['Voltage_1:']
vin = data2['Voltage_2:']

plt.plot(vin, vout, 'b.')
plt.title('Vout over Vin')
plt.xlabel('Vin (volts)')
plt.ylabel('Vout (volts)') 
plt.legend()
plt.grid()


slope = (np.mean(vout) / np.mean(vin))
print('Gain is equal to the slope of the graph which is', slope)
```

 # Exercise 1:
 
 For this part our goal is to find the probability that we will see a defect between t=4.5min and t=5.5min. So I used np.exp to calculate the CDF and graphed it as well as printed it out so it is clear what it is. The probabilty is 0.42305019 - 0.36237185 = 0.06067834    

```python
Tau=10
ax=plt.figure(figsize=(10,6))
t=np.linspace(4.5,5.5,100)      
CDF=1-np.exp(-t/Tau)           

plt.plot(t,CDF,'b-')
plt.xlabel('Time (milliseconds)')
plt.ylabel(r'Cumulative Distribution Function')
plt.title("CDF of the of time between defects")
plt.grid()
```

  To prove that the expectation value is equal to tau you have to integrate which I have shown down below. Once with just tau and then with the value I have for tau plugged in.
 
$$\int_{0}^{\infty} \frac{te^{-t/\tau }}{-\tau} = (-e^{\frac{-\infty}{\tau}} * (\infty + \tau)) - (-e^{0} * \tau) = \tau $$
$$ \int_{0}^{\infty} \frac{te^{-t/10}}{-10} = (-e^{\frac{-\infty}{10}} * (\infty + 10)) - (-e^{0} * 10) = 10 $$
